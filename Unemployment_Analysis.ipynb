# Unemployment Analysis with Python

## Project Objective

The objective of this project is to analyze unemployment trends across different regions of India using Python. The analysis focuses on identifying patterns in unemployment rates, employment levels, labour participation rates, and the impact of COVID-19 on unemployment.

## Import Libraries

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

## Load Dataset

```python
df = pd.read_csv('Unemployment_Rate_upto_11_2020.csv')

df.head()
```

## Data Exploration

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

## Data Cleaning

```python
df.columns = df.columns.str.strip()

df.columns
```

```python
df['Date'] = pd.to_datetime(
    df['Date'].str.strip(),
    dayfirst=True
)

df['Date'].head()
```

## Exploratory Data Analysis (EDA)

```python
plt.figure(figsize=(14,6))

region_unemployment = (
    df.groupby('Region')
    ['Estimated Unemployment Rate (%)']
    .mean()
    .sort_values(ascending=False)
)

region_unemployment.plot(kind='bar')

plt.title(
    'Average Unemployment Rate by Region'
)

plt.ylabel(
    'Unemployment Rate (%)'
)

plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    data=df,
    x='Date',
    y='Estimated Unemployment Rate (%)'
)

plt.title(
    'Unemployment Rate Over Time'
)

plt.show()
```

```python
plt.figure(figsize=(10,5))

sns.histplot(
    df['Estimated Employed'],
    kde=True
)

plt.title(
    'Distribution of Employment'
)

plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='Estimated Labour Participation Rate (%)',
    y='Estimated Unemployment Rate (%)'
)

plt.title(
    'Labour Participation vs Unemployment Rate'
)

plt.show()
```

```python
covid_data = df[
    df['Date'] >= '2020-03-01'
]

plt.figure(figsize=(12,6))

sns.lineplot(
    data=covid_data,
    x='Date',
    y='Estimated Unemployment Rate (%)'
)

plt.title(
    'Unemployment During COVID-19 Period'
)

plt.show()
```

## Key Insights

### Findings

1. Unemployment rates increased significantly during the COVID-19 lockdown period.

2. Employment distribution was uneven across regions.

3. Labour participation rate showed a weak relationship with unemployment.

4. Certain regions experienced consistently higher unemployment than others.

5. Unemployment peaked during the pandemic and gradually declined afterwards.
